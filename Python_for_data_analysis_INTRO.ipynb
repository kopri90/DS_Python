````python
import json
path = 'C:\Users\Wojtek Biedzinski\Documents\MK PYTHON\UDACITY Python for data analysys\JUPYTER\usagov_bitly_data2012-03-16-1331923249.txt'
records = [json.loads(line) for line in open(path)]

records[0]
records[0]['u']
````

````python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]

time_zones[:10]
len(time_zones)
````

````python
# using Pandas dataFrame
from pandas import DataFrame, Series
import pandas as pd
frame = DataFrame(records)

frame
````

````python
tz_counts = frame['tz'].value_counts()
#daje 10 pierwszy, czyli najwyższych wyników
tz_counts[:10]
````

````python
# Cleaning data from na, or empty value
clean_tz = frame['tz'].fillna('Missing')
clean_tz[clean_tz == ''] = 'Unknown'
tz_counts = clean_tz.value_counts()
tz_counts[:10]
tz_counts['Missing']
````

````python
#Visualization of timez zones
%matplotlib inline
tz_counts[:10].plot(kind='barh', rot=0)
````

````python
# a [second column] field contains information about the browser, device, or application used to perform the URL shortening
frame['a'][1]
frame['a'][3438]
results = Series([x.split()[0] for x in frame.a.dropna()])
results
````

````python
#usuwanie null wartości
cframe = frame[frame.a.notnull()]

operating_system = np.where(cframe['a'].str.contains('Windows'), 'Windows', 'Not Windows')
by_tz_os = cframe.groupby(['tz', operating_system])

by_tz_os

agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts 
````

````python
indexer = agg_counts.sum(1).argsort()
indexer
count_subset = agg_counts.take(indexer)[-10:]
````

````python
count_subset.plot(kind='barh')
````

````python
count_subset.plot(kind='barh', stacked=True)
````

````python
normed_subset = count_subset.div(count_subset.sum(1), axis=0)
count_subset
normed_subset
normed_subset.plot(kind='barh', stacked=True)
````
